```python
#Code task 5#
#Count missing values in number and percetage and present the result in a single table
missing_values = pd.concat([cust_churn.isnull().sum(), 100*cust_churn.isnull().mean()],axis=1)
missing_values.columns =['count','%']
missing_values.sort_values(by="%", ascending=False)
```

2.5 Number of churn by gender

```python
gender_churn_sum = pd.concat([cust_churn.groupby('Gender')['Gender'].count(),cust_churn.groupby('Gender')['Churn Value'].sum()],axis=1)
gender_churn_sum["%"]=100*gender_churn_sum['Churn Value']/gender_churn_sum['Gender']
gender_churn_sum
```

```python
mean_tenure = pd.Series(cust_churn['Tenure Months'].mean(), name='Mean')
median_tenure = pd.Series(cust_churn['Tenure Months'].median(), name='Median')
max_tenure = pd.Series(cust_churn['Tenure Months'].max(), name='Max')
quantile_75 = pd.Series(cust_churn['Tenure Months'].quantile(0.75), name='Q(75)')
quantile_25 = pd.Series(cust_churn['Tenure Months'].quantile(0.25), name='Q25')
# Concatenate Series into a DataFrame
churn_stat = pd.concat([mean_tenure, median_tenure, max_tenure,quantile_75,quantile_25], axis=1)

print(churn_stat)
```
